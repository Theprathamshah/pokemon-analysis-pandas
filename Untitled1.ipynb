# Find the outlier

Take a look at the scatterplot that correlates pokemons Attack and Defense. What's the Name of the pokemon that is a clear outlier (strong Defense, but very low Attack).

Start project before checking your activities.


```python
plt.figure(figsize=(8,6))
plt.scatter(df["attack"], df["defense"], alpha=0.6)
plt.xlabel("Attack")
plt.ylabel("Defense")
plt.title("Attack vs Defense")
plt.show()

attack_threshold = df["attack"].quantile(0.05)
defense_threshold = df["defense"].quantile(0.95)

print(attack_threshold,defense_threshold)
outlier = df[
    (df["attack"] <= attack_threshold) &
    (df["defense"] >= defense_threshold)
][["name", "attack", "defense"]]

outlier
```

# How many Fire-Flying Pokemons are there?


```python
fire_flying = df[
    (df["type1"] == "fire") &
    (df["type2"] == "flying")
]["name"]
print(len(fire_flying))
```

# How many 'Poison' pokemons are across both types?


```python
poison_pokemon = df[
    (df["type1"] == "poison") |
    (df["type2"] == "poison")
]["name"]
print(len(poison_pokemon))
```

# Name the pokemon of Type 1 Ice which has the strongest defense?


```python
strong_ice_defense = df[df["type1"]=="ice"].sort_values(by="defense", ascending=False).iloc[0]["name"]
print(strong_ice_defense)
```

# What's the most common type of Legendary Pokemons?


```python
legendary_type_counts = (
    df[df["is_legendary"] == 1]["type1"]
    .value_counts()
)

print(legendary_type_counts)

legendary_type_counts.idxmax(), legendary_type_counts.max()
```

# Find the most powerful pokemon from the first 3 generations, of type water ?


```python
water_gen1_3 = df[
    (df["type1"] == "water") &
    (df["generation"].isin([1, 2, 3]))
]

# get most powerful (highest Total)
most_powerful = water_gen1_3.sort_values(by="base_total", ascending=False).iloc[0]

print(most_powerful[["name", "base_total", "generation"]])
```

# Take a look at the distribution of Pokemon's speed in the histogram included in the notebook. The red lines separate the slowest (bottom 5%) and fastest (top 5%) pokemons.

# Select those pokemons that are either very slow (with Speed below the bottom 5%) or very fast (Speed above top 95%).

# Store your results in the variable slow_fast_df.


```python
slowest_mark = df["speed"].quantile(0.05)
fastest_mark = df["speed"].quantile(0.95)
print(slowest_mark, fastest_mark)
slow_fast_df = df[(df["speed"]>=fastest_mark) | (df["speed"]<=slowest_mark)][["name","speed"]].sort_values(by="speed")
slow_fast_df
```
